## FEATURE EXTRACTION FROM MUSIC SIGNALS

```python
import numpy, pandas, librosa, sklearn
import config, functions
import copy
```

```python
float_formatter = lambda x: "%.3f" % x
numpy.set_printoptions(formatter={'float_kind':float_formatter})
```

```python
DATASET_DIRECTORY=config.CreateDataset.DATASET_DIRECTORY
```

```python
path_of_songs=librosa.util.find_files(DATASET_DIRECTORY)
```

```python
number_of_songs=config.CreateDataset.Number_of_songs
number_of_genres=config.CreateDataset.Number_of_genres
```

```python
sample_rate=config.CreateDataset.Sampling_rate
songs=[]

for song in path_of_songs:
    x,sr=librosa.load(song,sr=sample_rate,duration=5.0)
    songs.append(x)
    
print("DONE !")
```

```python
extract_features=copy.deepcopy(functions.extract_features)

frame_size=2048
hop_size=512

features=numpy.array([extract_features(x,sample_rate,frame_size,hop_size) for x in songs])


print("DONE !")
```

```python
print("Shape of feature table: {}\n".format(features.shape))

print("An overview of the feature table:\n")
print(features[1,:])
```

```python
scaler = sklearn.preprocessing.MinMaxScaler(feature_range=(-1, 1))
features = scaler.fit_transform(features)
```

```python
create_label_column=copy.deepcopy(functions.create_label_column)
y_data=create_label_column(number_of_songs, number_of_genres)

Feature_Names=['meanZCR','stdZCR','meanSpecCentroid','stdSpecCentroid','meanSpecContrast','stdSpecContrast',
            'meanSpecBandwidth','stdSpecBandwidth','meanSpecRollof','stdSpecRollof',
          'meanMFCC_1','stdMFCC_1','meanMFCC_2','stdMFCC_2','meanMFCC_3','stdMFCC_3',
           'meanMFCC_4','stdMFCC_4','meanMFCC_5','stdMFCC_5','meanMFCC_6','stdMFCC_6',
             'meanMFCC_7','stdMFCC_7','meanMFCC_8','stdMFCC_8','meanMFCC_9','stdMFCC_9',
            'meanMFCC_10','stdMFCC_10','meanMFCC_11','stdMFCC_11','meanMFCC_12','stdMFCC_12',
             'meanMFCC_13','stdMFCC_13'
             ]

data_set=pandas.DataFrame(features,columns=Feature_Names)
data_set['Genre']=y_data

data_set[:10].style
```

```python
data_set.to_csv("data_set.csv",index=False)
```
